__<H2>Piechart__

__Starting at 0Â°__

```python
import matplotlib.pyplot as plt
 
y =[35, 25, 45, 15]
mylabels = ["C", "C++", "Python", "java"]
 
plt.pie(y, labels = mylabels)
plt.show()
```

```python
 import matplotlib.pyplot as plt
import numpy as np
 
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
 
plt.pie(y, labels = mylabels)
plt.show()
```

```python
import matplotlib.pyplot as plt
y = [35, 25, 25, 15]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
 
plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

__Legend__

```python
import matplotlib.pyplot as plt
 
y =[35, 25, 45, 15]
mylabels = ["C", "C++", "Python", "java"]
 
plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

```python
 import matplotlib.pyplot as plt
 
y =[35, 25, 45, 15]
mylabels = ["C", "C++", "Python", "java"]
 
plt.pie(y, labels = mylabels)
plt.legend(title="Coding")
plt.show()
```

 __Explode__

```python
 import matplotlib.pyplot as plt
 
y =[35, 25, 45, 15]
mylabels = ["C", "C++", "Python", "java"]
x=[.2,0,0,0]
plt.pie(y, labels = mylabels, explode=x)
plt.show()
```

```python
import matplotlib.pyplot as plt
 
y =[35, 25, 45, 15]
mylabels = ["C", "C++", "Python", "java"]
x=[.1,.1,.1,.1]
plt.pie(y, labels = mylabels, explode=x)
plt.show()
```
